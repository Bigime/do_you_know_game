```python
!pip install jamo

from jamo import h2j, j2hcj

def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python

def make_user_dic_csv(morpheme_type, word_list, user_dic_file_name):
  file_data = []

  for word, score in word_list:
    jongsung_TF = get_jongsung_TF(word)

    line = f"{word},,,{score},{morpheme_type},*,{jongsung_TF},{word},*,*,*,*,*\n"

    file_data.append(line)

  with open("./user-dic/user-nnp.csv", 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)
```

## [참고] 동시 등장 행렬 (Co-occurence Matrix)

의미 연결망 분석(Semantic Network Analysis)에 자주 활용되는 **동시 등장 행렬**을 계산하는 방법을 알아보도록 하겠습니다.

여기서는 단어 카운트 벡터간의 내적을 계산하여 동시 등장 빈도를 계산하는데, 그 원리를 간단하게 알아봅니다.

```python
# 아래와 같은 예제를 살펴보겠습니다.
# 단어 토큰을 중심으로 문서별 등장 횟수를 리스트로 표현했습니다.
import pandas as pd

example_df = pd.DataFrame({'허미미' : [1, 1, 1, 0, 1],
                            '유도' : [1, 1, 0, 1, 1],
                            '금메달' : [1, 0, 1, 1, 0],
                            '독립' : [0, 1, 1, 1, 1]})

example_df
```

```python
# 이를 전치(Transpose)하면 문서별 단어 등장 횟수를 살펴볼 수 있습니다.
example_df.T
```

이 과정에서 각 항목(열)이 다른 항목과 얼마나 자주 함께 등장하는지를 계산합니다. 예를 들어, `허미미`와 `유도`가 함께 등장하는 횟수는 두 벡터의 내적을 통해 계산됩니다. 이 둘을 내적한다는 의미가 무엇일까요?

예시처럼 `허미미` 벡터가 `[1, 1, 1, 0, 1]`이고, `유도` 벡터가 `[1, 1, 0, 1, 1]`이라면, 두 벡터의 내적은 `1*1 + 1*1 + 1*0 + 0*1 + 1*1 = 3`이 됩니다. 즉, 허미미와 유도가 3번 함께 등장했음을 의미합니다. 이렇게 한 문서에서 같이 등장했을 경우, 문서별로 각 단어의 등장 횟수가 곱해집니다. 이를 모두 더하게 되면 전체 문서에 대한 동시 등장 빈도를 계산할 수 있습니다.

```python
import numpy as np
pd.DataFrame(np.dot(example_df.values.T, example_df.values), columns = example_df.columns, index = example_df.columns)
```

### 파리올림픽 뉴스 기사로 동시 등장 행렬 계산해보기

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(tokenizer = lambda x: [word for word in mecab.nouns(x) if len(word) != 1],
                             min_df = 20, max_features = 100,
                             stop_words  = ['기자', '파리', '올림픽', '선수'] + ['프랑스', '대회', '경기', '시간', '한국', '이번', '대표'])
# 한 글자 단어는 제거했습니다.
# 이 경우에는 최대 단어 개수인 max_features 의 값을 설정해주는 것이 좋습니다.

count_matrix = vectorizer.fit_transform(df['title'])
# 약 1분 소요
```

```python
word_df = pd.DataFrame(count_matrix.todense(), columns = vectorizer.get_feature_names_out()); word_df
```

```python
word_df = word_df.where(word_df <= 1, 1); word_df
```

```python
co_matrix = np.dot(word_df.T, word_df)
```

```python
feature_names = vectorizer.get_feature_names_out()
co_df = pd.DataFrame(co_matrix, index = feature_names, columns = feature_names)
```

```python
co_df
```

```python
co_df_lower = co_df.where(np.tril(np.ones(co_df.shape), k = -1).astype(bool))
co_df_lower
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# LabelEncoder 객체 생성
le = LabelEncoder()

# columns에 대해 Label Encoding 수행
columns = co_df.columns
encoded_columns = le.fit_transform(columns)

# ID와 Label이 포함된 데이터프레임 생성
encoded_df = pd.DataFrame({'Id': encoded_columns, 'Label': columns})

# Node 정보 추출하기
encoded_df.to_csv('파리올림픽 2024 동시 등장 행렬_Node.csv', index = False)
```

```python
# 데이터프레임을 재구조화
melted_df = co_df_lower.reset_index().melt(id_vars='index', var_name='target', value_name='weight')

# 'index' 컬럼명을 'source'로 변경
melted_df.rename(columns={'index': 'source'}, inplace=True)

# 컬럼명을 source / target / weight 로 변경하는 이유는 Gephi에서 활용하기 위함입니다.
```

```python
# 중간 결과 예시 확인
melted_df
```

```python
# 자기 자신을 카운트한 경우 제거하기
melted_df.loc[melted_df['source'] == melted_df['target'], 'weight'] = np.nan
melted_df.dropna(inplace = True)
```

```python
# 동시 등장 빈도 순으로 정렬하기
melted_df = melted_df.sort_values(by = 'weight', ascending = False).copy()
```

```python
# 동시 등장 빈도가 0인 행은 제거하기
melted_df = melted_df[melted_df['weight'] != 0].copy(); melted_df
```

```python
# 최종 결과 예시 확인
melted_df.head(50)
```

```python
# 동일한 LabelEncoder 사용 (이미 fit된 le를 사용)
melted_df['source'] = le.transform(melted_df['source'])
melted_df['target'] = le.transform(melted_df['target'])

# Edge 정보 추출하기
melted_df.to_csv('파리올림픽 2024 동시 등장 행렬_Edge.csv', index = False)
```
